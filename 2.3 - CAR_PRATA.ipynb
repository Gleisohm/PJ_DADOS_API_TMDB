### Carga da Tabela Prata da Arquitetura Medallion

```
%sql
USE DBTMDB;
--validando tipo das colunas da tabela delta bronze
DESCRIBE EXTENDED TMDB_BRONZE;
```

```
%sql
--usando a função explode do SparkSQL para transformar cada posição do array da lista de filmes em uma linha na tabela
SELECT page, explode(results), total_pages, total_results, data_cadastro, user_cadastro
FROM TMDB_BRONZE;
```

```
%sql  
--criando view temporária para preparar os dados e popular as linhas com elementos do array
USE DBTMDB;

CREATE OR REPLACE GLOBAL TEMP VIEW TMDB_PRE_PRATA_VW AS 
SELECT page, explode(results), total_pages, total_results, data_cadastro, user_cadastro
FROM TMDB_BRONZE;

SELECT * FROM GLOBAL_TEMP.TMDB_PRE_PRATA_VW;
```

```
%sql
--A nova view tem a coluna col como struct, permitindo acesso aos objetos internos com . ou :
DESCRIBE GLOBAL_TEMP.TMDB_PRE_PRATA_VW;
```

```
%sql
/*Ajustando dados para criar tabela prata:
coluna RELEASE_DATE ajustada para o tipo data 
coluna ADULT com tratamento para indicação do conteúdo
coluna POPULARITY com tratamento de casas decimais e tipo float
coluna VOTE_AVERAGE com tratamento tipo float
coluna POSTER_PATH com tratamento para link completo da imagem 
Junção com tabela estática TMDB_GENEROS_REF para recuperar a descrição (NAME) dos gêneros. Aqui a coluna GENRE_IDS é um array, então usei a função array_contains() para buscar colunas correspondentes, porém para cada gênero será gerada uma linha. Ex.: (Tabela TMDB_PRE_PRATA_VW tem COL.GENRE_IDS = 2 e 3, será gerada uma linha para o COL.GENRE_IDS = 2 e uma linha para o COL.GENRE_IDS = 3, com suas respectivas descrições de gênero)
*/

SELECT 
COL.TITLE,
COL.ORIGINAL_TITLE,
CASE WHEN col.adult = false THEN "Não" ELSE "Sim" END as ADULT, 
COL.GENRE_IDS,
B.NAME, --coluna recuperada na tabela TMDB_GENEROS_REF
CAST(format_number(COL.POPULARITY,2) AS FLOAT) as POPULARITY,
to_date(COL.RELEASE_DATE, 'yyyy-MM-dd') as RELEASE_DATE,
CAST(COL.VOTE_AVERAGE AS FLOAT) as VOTE_AVERAGE,
COL.VOTE_COUNT,
concat('https://image.tmdb.org/t/p/original', COL.POSTER_PATH) as POSTER_PATH,
COL.OVERVIEW,
DATA_CADASTRO,
USER_CADASTRO
FROM GLOBAL_TEMP.TMDB_PRE_PRATA_VW A
INNER JOIN TMDB_GENEROS_REF B
ON array_contains(COL.GENRE_IDS, B.ID); -- aqui também seria possível utilizar a função explode
--ON  COL.GENRE_IDS = B.ID aqui seria um join comum, mas não daria certo pq COL.GENRE_IDS é um array
```

```
%sql
--criando segunda view temporária a partir da última query para preparar os dados para carregar na tabela prata. 
--Aqui as linhas estão repetidas devido ao desmembramento dos gêneros da última query, que estão em array. Na próxima query iremos transformar a coluna NAME em um array, assim como é a coluna GENRE_IDS
USE DBTMDB;

CREATE OR REPLACE GLOBAL TEMP VIEW TMDB_PRE_PRATA_VW2 AS 
SELECT 
COL.TITLE,
COL.ORIGINAL_TITLE,
CASE WHEN col.adult = false THEN "Não" ELSE "Sim" END as ADULT, 
COL.GENRE_IDS,
B.NAME, --coluna recuperada na tabela TMDB_GENEROS_REF
CAST(format_number(COL.POPULARITY,2) AS FLOAT) as POPULARITY,
to_date(COL.RELEASE_DATE, 'yyyy-MM-dd') as RELEASE_DATE,
CAST(COL.VOTE_AVERAGE AS FLOAT) as VOTE_AVERAGE,
COL.VOTE_COUNT,
concat('https://image.tmdb.org/t/p/original', COL.POSTER_PATH) as POSTER_PATH,
COL.OVERVIEW,
DATA_CADASTRO,
USER_CADASTRO
FROM GLOBAL_TEMP.TMDB_PRE_PRATA_VW A
INNER JOIN TMDB_GENEROS_REF B
ON array_contains(COL.GENRE_IDS, B.ID);

SELECT * FROM GLOBAL_TEMP.TMDB_PRE_PRATA_VW2;
```

```
%sql
DESCRIBE GLOBAL_TEMP.TMDB_PRE_PRATA_VW2;
```

```
%sql
/* usando a função COLLECT_LIST(B.NAME) para agrupar os gêneros (NAME) pela coluna ID. Agora temos a coluna GENRE_IDS e também a coluna GENERO recuperada na tabela de referência, mas agora ela está em um array também.
coluna ID utilizando a função monotonically_increasing_id()+1 para gerar um ID único. */

SELECT 
monotonically_increasing_id()+1 as ID, 
TITLE, 
ORIGINAL_TITLE, 
GENRE_IDS,
collect_list(NAME) AS GENERO, 
ADULT,
POPULARITY,
RELEASE_DATE,
VOTE_AVERAGE,
VOTE_COUNT,
POSTER_PATH,
OVERVIEW,
DATA_CADASTRO,
USER_CADASTRO
FROM GLOBAL_TEMP.TMDB_PRE_PRATA_VW2 
GROUP BY TITLE, ORIGINAL_TITLE, GENRE_IDS, ADULT, POPULARITY, RELEASE_DATE, VOTE_AVERAGE, VOTE_COUNT, POSTER_PATH, OVERVIEW, DATA_CADASTRO, USER_CADASTRO
ORDER BY ID;

```

```
%sql
USE DBTMDB;
--criando tabela prata com todas as transformações realizadas em Spark SQL

CREATE OR REPLACE TABLE TMDB_PRATA AS
SELECT 
monotonically_increasing_id()+1 as ID,
TITLE, 
ORIGINAL_TITLE, 
GENRE_IDS,
collect_list(NAME) AS GENERO, 
ADULT,
POPULARITY,
RELEASE_DATE,
VOTE_AVERAGE,
VOTE_COUNT,
POSTER_PATH,
OVERVIEW,
DATA_CADASTRO,
USER_CADASTRO
FROM GLOBAL_TEMP.TMDB_PRE_PRATA_VW2 
GROUP BY TITLE, ORIGINAL_TITLE, GENRE_IDS, ADULT, POPULARITY, RELEASE_DATE, VOTE_AVERAGE, VOTE_COUNT, POSTER_PATH, OVERVIEW, DATA_CADASTRO, USER_CADASTRO;

```

```
%sql
SELECT * FROM TMDB_PRATA
ORDER BY 1;
```

```
%sql
DESCRIBE EXTENDED TMDB_PRATA; 
```
